#### 12. Set Operations
Given two sets:<br>
a = {1, 2, 3, 4},
b = {3, 4, 5, 6}<br>
<br>
Find the union<br>
Find the intersection<br>
Find the difference a - b and b - a



```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(f"Union: {a | b}")
print(f"Intersection: {a & b}")
# Elements in a but not in b
print(f"Difference (a - b): {a - b}")
# Elements in b but not in a
print(f"Difference (b - a): {b - a}")
```

#### 15. Nested Dictionaries
Create a dictionary to store information about two students:<br>

students = {
    'Ava': {'age': 20, 'major': 'CS'},
    'Leo': {'age': 22, 'major': 'Math'}
}<br>
- Print Ava's major<br>
- Add a new student<br>
- Change Leo’s age to 23

```python
students = {
    'Ava': {'age': 20, 'major': 'CS'},
    'Leo': {'age': 22, 'major': 'Math'}
}

print(f"Ava's major: {students['Ava']['major']}")

students['Sneha'] = {'age': 21, 'major': 'CS'}
print(f"After adding new students: {students}")

students['Leo']['age'] = 23
print(f"After changing Leo's age: {students}")
```

#### 16. Write a program that:

- Takes a sentence from the user<br>
- Splits it into words<br>
- Creates a dictionary with the frequency of each word

```python
sentence = input("Enter any sentence: ")

words = sentence.split()
print(words)

dic_frequency = {}
for word in words:
    if word in dic_frequency:
        dic_frequency[word] +=1
    else:
        dic_frequency[word] = 1
        
print(dic_frequency)
```
