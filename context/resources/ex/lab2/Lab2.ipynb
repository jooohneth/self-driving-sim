## Main Difference: Color Channel Order
# OpenCV (cv2.imshow):
- Displays images in BGR format (Blue, Green, Red)
- Shows the image with correct colors automatically
- No conversion needed


# Matplotlib (plt.imshow):

- Without conversion, colors look wrong (blue becomes red, red becomes blue)
- Must convert using cv2.cvtColor(image, cv2.COLOR_BGR2RGB) to display correctly

9. Having an image, create a 1x3 subplot that shows:

- Original RGB image
- Red channel
- HSV version
- Adjust the figure size for better visualization.

```python
import cv2
import matplotlib.pyplot as plt

cat = cv2.imread('Cat.jpg')
img = cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)

# Extract red channel (index 0 in RGB)
red_channel = img[:, :, 0] 

# HSV version
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Subplot 1: Original RGB
plt.subplot(131)  # 1 row, 3 columns, position 1
plt.imshow(img)
plt.title('Original RGB')
plt.axis('off')

# Subplot 2: Red Channel
plt.subplot(132)  # 1 row, 3 columns, position 2
plt.imshow(red_channel, cmap='Reds')  # or cmap='Reds' for red tint
plt.title('Red Channel')
plt.axis('off')

# Subplot 3: HSV Version
plt.subplot(133)  # 1 row, 3 columns, position 3
plt.imshow(hsv)
plt.title('HSV')
plt.axis('off')

plt.tight_layout()  # Adjusts spacing between subplots
plt.show()
```

#### 12. Convolution
- Apply this kernel on an image.

         [1, 0, -1],
         [1, 0, -1],
         [1, 0, -1] 

- Use matplotlib to show your original and convolved images side by side.
- Describe the observed effect of the convolution operation. What features does this kernel highlight?

```python
import numpy as np
import cv2

cat = cv2.imread('kiki.png')
cat= cv2.cvtColor(cat, cv2.COLOR_BGR2RGB)

kernel = np.array([
        [1, 0, -1],
         [1, 0, -1],
         [1, 0, -1] ])

edge = cv2.filter2D(cat, cv2.CV_8U, kernel)

# cv2.imshow('edge', edge)
# cv2.waitKey(100)
# cv2.destroyAllWindows()
```

#### 13. Denoising
- Add noise to your image.
- Apply median filtering to denoise an image.

```python
import numpy as np
import cv2

cat = cv2.imread('kiki.jpg')

mean = 0
std = 60

cat.dtype
```

```python
noise = np.random.normal(mean, std, cat.shape).astype('float32')
```

```python
noise.dtype
```

```python
noisy_image = cv2.add(cat.astype('float32'), noise)
noisy_image = np.clip(noisy_image, 0, 255).astype('uint8')

# Apply median filtering to denoise
denoised_image = cv2.medianBlur(noisy_image, 5)
```

```python
cv2.imshow('frame', noisy_image)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

```python
cv2.imshow('frame', denoised_image)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

```python
lucy = cv2.imread('Cat.jpg')

mean = 0
std = 20
noise = np.random.normal(mean, std, lucy.shape).astype('float32')
noisy_image = cv2.add(lucy.astype('float32'), noise)
noisy_image = np.clip(noisy_image, 0, 255).astype('uint8')
```

```python
# Denoise using median filter
# The 5 in cv2.medianBlur(noisy_image, 5) is the kernel size (also called window size ). 
# The median filter looks at a neighborhood of pixels around each pixel and replaces it with the median value of that neighborhood.
# Kernel size = 5 means it looks at a 5×5 pixel window (25 pixels total)
# It sorts those 25 pixel values and picks the middle one
denoised = cv2.medianBlur(noisy_image, 5)

cv2.imshow('noisy', noisy_image)
cv2.imshow('denoised', denoised)
cv2.waitKey(100)
cv2.destroyAllWindows()
```
