### For np.zeros we have to include dtype. For np.zeros_like it's not mandatory

```python
new_image = np.zeros(im.shape, dtype ='uint8')
```

```python
new_image[:, :, 0]= im[:, :, 0]
```

```python
cv2.imshow('image1',new_image)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

* cv2.BORDER_CONSTANT (Adds a border filled with a solid color)
* cv2.BORDER_REFLECT (Adds a border by reflecting/mirroring the image edges)

```python
padded_img = cv2.copyMakeBorder(im, 20, 20, 10, 0, cv2.BORDER_CONSTANT, value = (255,255,255))
```

```python
cv2.imshow('image1',padded_img)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```

```python
padded_img1 = cv2.copyMakeBorder(im, 40, 40, 40, 40, cv2.BORDER_REFLECT)
```

```python
cv2.imshow('image1',padded_img1)
cv2.waitKey(1000)
cv2.destroyAllWindows()
```
