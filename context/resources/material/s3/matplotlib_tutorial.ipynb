# Introduction to Matplotlib
Matplotlib is a powerful plotting library for creating static, animated, and interactive visualizations in Python.

In this notebook, we'll cover:
- Basic plotting
- Customizing plots
- Subplots
- Plotting from NumPy

```python
import matplotlib.pyplot as plt
```

### Basic plotting

```python
x = [1, 2, 3, 4, 5]
y = [i**2 for i in x]

plt.plot(x, y)
plt.title("Basic Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

### Plot costumization

```python
plt.plot(x, y, color='red', marker='o', linestyle='--', linewidth=2, markersize=8)
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

### Multiple plots

```python
x = [1, 2, 3, 4, 5]
y1 = [i for i in x]

y2 = [i**2 for i in x]
y3 = [i**3 for i in x]

plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = x^2")
plt.plot(x, y3, label="y = x^3")

plt.legend()
plt.title("Multiple Lines")
plt.show()
```

### Subplots

```python
fig, axs = plt.subplots(1, 2)
axs[0].plot(x, y1)
axs[0].set_title("y = x")

axs[1].plot(x, y2)
axs[1].set_title("y = x^2")

plt.tight_layout()
plt.show()


### OR

plt.subplot(121)
plt.title("y = x")
plt.plot(x, y1)

plt.subplot(122)
plt.title("y = x^2")
plt.plot(x, y2)


plt.savefig("metrics.png")
plt.show()
```

### Bar charts and histograms

```python
categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title("Histogram")
plt.show()
```

### Pie charts

```python
labels = ['Apples', 'Bananas', 'Cherries']
sizes = [30, 45, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Fruit Pie Chart")
plt.show()
```
