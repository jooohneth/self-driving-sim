```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("cat2.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
constant = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_CONSTANT, value=[50, 50, 200])
reflect = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REFLECT)
```

In more detail: `plt.subplot(221)` means “make a grid of 2 rows and 2 columns, and now select position 1” (top‑left). 

```python
plt.subplot(221)
plt.imshow(constant)

plt.subplot(222)
plt.imshow(reflect)

plt.subplot(223)
plt.imshow(reflect)

plt.subplot(224)
plt.imshow(reflect)
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
```

`image`: the input image (NumPy array in OpenCV’s BGR format). The function modifies this image in-place and also returns it.

- `(50, 50)`: starting point of the line (x=50, y=50) in pixel coordinates.

- `(200, 600)`: ending point of the line (x=200, y=600).

- `(0, 0, 0)`: color of the line in BGR. This is black. For red, you’d use `(0, 0, 255)`.

- `3`: thickness of the line in pixels.



```python
kiki_line = cv2.line(image, (50, 50), (200, 600), (0, 0, 0), 3)
```

```python
cv2.imshow('frame', kiki_line)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

Creates a white frame

```python
new_img = np.ones((300, 300), dtype='uint8') * 255
cv2.imshow('frame', new_img)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

```python
rect = cv2.rectangle(new_img, (50,50), (250, 250), (0, 0, 0), 5)
cv2.imshow('frame', rect)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

```python
circle = cv2.circle(new_img, (150, 150), 50, (60, 80, 100), 5)
cv2.imshow('frame', circle)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

```python
text = cv2.putText(image, "Geometry", (10, 200), cv2.FONT_HERSHEY_SIMPLEX, 3, (0, 0, 255), 2)
cv2.imshow('frame', text)
cv2.waitKey(100)
cv2.destroyAllWindows()
```

```python
original_fps = cap.get(cv2.CAP_PROP_FPS)
desired_fps = 10
counter = 0

while True:
    _, frame = cap.read()
    frame = cv2.resize(frame, (800, 500))
    # cv2.imshow('frame', frame)
    # cv2.waitKey(100)
    
    if counter % desired_fps == 0:
        cv2.imshow('frame', frame)
        if cv2.waitKey(10) == ord('q'):
            break
    
    counter += 1
    
cap.release()
cv2.destroyAllWindows()
```
