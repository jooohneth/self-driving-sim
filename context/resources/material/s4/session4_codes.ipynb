```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("../Lucy.jpg")
# image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
constant = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_CONSTANT, value=[150, 70, 50])
reflect = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REFLECT)
replicate = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_REPLICATE)
wrap = cv2.copyMakeBorder(image, 50, 50, 50, 50, cv2.BORDER_WRAP)
```

```python
plt.subplot(221)
plt.imshow(constant)
plt.title("constant")

plt.subplot(222)
plt.imshow(reflect)

plt.subplot(223)
plt.imshow(replicate)

plt.subplot(224)
plt.imshow(wrap)

plt.show()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
```

```python
lucy_line = cv2.line(image, (50, 50), (200, 600), (0, 0, 0), 3)
```

```python
cv2.imshow('frame', lucy_line)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_img = np.ones((300, 300, 3), dtype='uint8') * 255
cv2.imshow('frame', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
rect = cv2.rectangle(new_img, (50, 50), (200, 150), (0, 255, 0), 5)
cv2.imshow('frame', rect)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.circle(new_img, (150, 150), 50, (60, 80, 100), 5)
cv2.imshow('frame', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.putText(new_img, "Computer Vision", (10, 200), cv2.FONT_HERSHEY_DUPLEX, 1, (0,0,255), 2)
cv2.imshow('frame', new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# inplace

cap = cv2.VideoCapture("S10_cars_video (1).mp4")
```

```python
original_fps = cap.get(cv2.CAP_PROP_FPS)
desired_fps = 10
counter = 0


while True:
    _, frame = cap.read()
    frame = cv2.resize(frame, (800, 500))
    # cv2.imshow('frame', frame)
    # cv2.waitKey(100)

    if counter % desired_fps == 0:
        cv2.imshow('frame', frame)
        if cv2.waitKey(10) == ord('q'):
            break

    
    counter += 1

cap.release()
cv2.destroyAllWindows()
```
