```python
import cv2
import numpy as np
```

```python
cat = cv2.imread("cat2.jpg")
```

```python
angel = 45
scale = 2

h, w  = cat.shape[:2]

h = h // 2
w = w // 2
```

```python
h
```

```python
w
```

```python
# matrix = np.array([])
m = cv2.getRotationMatrix2D((w,h), angel, scale)
```

```python
rotate = cv2.warpAffine(cat, m, (w*2, h*2))
```

```python
cv2.imshow('frame', rotate)
cv2.waitKey(10)
cv2.destroyAllWindows()
```

## Noise


```python
cat.shape
```

```python
mean = 0
std = 50
```

```python
noise = np.random.normal(mean, std, cat.shape).astype('uint8')
```

```python
noise
```

```python
noise.dtype
```

```python
noise.min()
```

```python
noise.max()
```

```python
# noisy_image = cv2.add(cat.astype('float64'), noise)
# noisy_image = np.clip(noisy_image, 0, 255)

noisy_image = cv2.add(cat, noise)
noisy_image = np.clip(noisy_image, 0, 255).astype('uint8')
```

adding to unit8 will add more brightness
float will add brightness to but when convert back it gives the original
