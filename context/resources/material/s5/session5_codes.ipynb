```python
brighter = cv2.add(lucy, 50)
cv2.imshow('frame', brighter)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow('frame', cv2.subtract(lucy, 100))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
_, image = cv2.threshold(lucy, 127, 255, cv2.THRESH_TOZERO_INV)
cv2.imshow('frame', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
