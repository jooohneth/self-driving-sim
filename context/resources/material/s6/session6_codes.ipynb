```python
import cv2
import numpy as np
```

```python
lucy = cv2.imread("../Lucy.jpg")
```

```python
angel = 45
scale = 1

h, w = lucy.shape[:2]

h = h//2
w = w//2
```

```python
w
```

```python
# matrix = np.array([
    
# ])

M = cv2.getRotationMatrix2D((w, h), angel, scale=0.5)
```

```python
rotated = cv2.warpAffine(lucy, M, (w*2, h*2))
```

```python
cv2.imshow('frame', rotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# NOISE

```python
lucy.shape
```

```python
mean = 0
std = 60
```

```python
noise = np.random.normal(mean, std, lucy.shape).astype('float32')
```

```python
noise.dtype
```

```python
noise.min()
```

```python
noise.max()
```

```python
noisy_image = cv2.add(lucy.astype('float32'), noise)
```

```python
noisy_image.min()
```

```python
noisy_image.max()
```

```python
noisy_image = np.clip(noisy_image, 0, 255).astype('uint8')
```

```python
cv2.imshow('frame', noisy_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
