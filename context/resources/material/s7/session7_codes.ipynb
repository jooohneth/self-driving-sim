```python
import numpy as np
import cv2
```

```python
square = cv2.imread('square.png')
square = cv2.cvtColor(square, cv2.COLOR_BGR2GRAY)
```

```python
kernel = np.array([[-1], [1]])
```

```python
edge = cv2.filter2D(square, cv2.CV_8U, kernel)
```

```python
cv2.imshow('edge', edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lucy = cv2.imread('../Lucy.jpg')
lucy = cv2.cvtColor(lucy, cv2.COLOR_BGR2GRAY)
edge = cv2.filter2D(lucy, cv2.CV_8U, kernel)
cv2.imshow('edge', edge)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lucy = cv2.imread('../Lucy.jpg')

mean = 0
std = 60
noise = np.random.normal(mean, std, lucy.shape).astype('float32')
noisy_image = cv2.add(lucy.astype('float32'), noise)
noisy_image = np.clip(noisy_image, 0, 255).astype('uint8')
```

```python
kernel = np.ones((25,25)) * (1/625)
denoised = cv2.filter2D(noisy_image, cv2.CV_8U, kernel)

cv2.imshow('noisy', noisy_image)
cv2.imshow('denoised', denoised)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
